```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_excel('dataset/bulanan-ekspor-impor.xlsx')

# Membuat kolom baru "Neraca Perdagangan"
df["Neraca Perdagangan"] = df["Nilai-Ekspor"] - df["Nilai-Impor"]

# Mengubah kolom "Bulan-Tahun" menjadi tipe data datetime
df['Bulan-Tahun'] = pd.to_datetime(df['Bulan-Tahun'], format='%Y-%m-%d')

df.head()
```

```python
# Mengambil data satu tahun (2019) dari DataFrame
data_2019 = df.loc[df['Bulan-Tahun'].dt.year == 2019]
data_2020 = df.loc[df['Bulan-Tahun'].dt.year == 2020]
data_2021 = df.loc[df['Bulan-Tahun'].dt.year == 2021]
data_2022 = df.loc[df['Bulan-Tahun'].dt.year == 2022]
data_2023 = df.loc[df['Bulan-Tahun'].dt.year == 2023]
```

```python
# Membuat subset data untuk nilai surplus dan defisit
surplus_data19 = data_2019[data_2019['Neraca Perdagangan'] > 0]
defisit_data19 = data_2019[data_2019['Neraca Perdagangan'] < 0]

surplus_data20 = data_2020[data_2020['Neraca Perdagangan'] > 0]
defisit_data20 = data_2020[data_2020['Neraca Perdagangan'] < 0]

surplus_data21 = data_2021[data_2021['Neraca Perdagangan'] > 0]
defisit_data21 = data_2021[data_2021['Neraca Perdagangan'] < 0]

surplus_data22 = data_2022[data_2022['Neraca Perdagangan'] > 0]
defisit_data22 = data_2022[data_2022['Neraca Perdagangan'] < 0]

surplus_data23 = data_2023[data_2023['Neraca Perdagangan'] > 0]
defisit_data23 = data_2023[data_2023['Neraca Perdagangan'] < 0]
```

```python
# Mengatur ukuran plot
plt.figure(figsize=(15, 8))  # Ubah ukuran sesuai kebutuhan Anda

# Menggambar bar plot
plt.bar(surplus_data22['Bulan-Tahun'].dt.month, surplus_data22['Neraca Perdagangan'], label='Surplus', color='darkgreen')

# Memeriksa apakah ada data defisit sebelum menambahkan bar plot dan label
if len(defisit_data22) > 0:
    plt.bar(defisit_data22['Bulan-Tahun'].dt.month, defisit_data22['Neraca Perdagangan'], label='Defisit', color='yellow')

# Menambahkan label sumbu x dan y serta judul plot
plt.xlabel('Bulan')
plt.ylabel('Nilai Neraca Perdagangan')
plt.title('Bar Plot Neraca Perdagangan 2022')

# Mengatur tampilan sumbu x dengan nama bulan
bulan_labels = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
plt.xticks(range(1, 13), bulan_labels)

# Menampilkan legend hanya jika ada data defisit
if len(defisit_data22) > 0:
    plt.legend()
if len(surplus_data22) > 0:
    plt.legend()

# Menampilkan plot
plt.show()
```
